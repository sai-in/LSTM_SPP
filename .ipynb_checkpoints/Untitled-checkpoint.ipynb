```python
import pandas as pd
import numpy as np
import pandas_datareader as web
import datetime as dt
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10,6)
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler, MinMaxScaler, StandardScaler

from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRFRegressor

from sklearn.metrics import *
```

```python
# import tensorflow as tf
# from tensorflow import keras
```

```python
stocks = ['MA']
start = dt.datetime(2010, 1, 1)
end = dt.datetime(2020, 1, 1)
df = web.DataReader(stocks, 'yahoo', start, end)
```

```python
df.head()
```

```python
df.columns = ['Adj_close','Close','High','Low','Open','Volume']
df.head()
```

```python
df['Close'].plot(figsize=(12,8))
```

```python
train_df = df[:'2018']
test_df = df['2019':]
```

```python
test_df['Close'].plot()
```

```python
cols = ['High','Low','Open','Volume']
X_train = train_df[cols]
X_test = test_df[cols]
y_train = train_df['Close']
y_test = test_df['Close']
```

```python
scaler = MinMaxScaler()
scaler.fit_transform(X_train)
scaler.transform(X_test)
```

```python
xgbr = XGBRFRegressor()
xgbr.fit(X_train,y_train)
```

```python
pred = xgbr.predict(X_train)
```

```python
xgbr.score(X_test,y_test)
```

```python
plt.figure(figsize=(12,8))
X_axis = list(range(len(y_train)))
plt.plot(X_axis,pred,c='r')
plt.plot(X_axis,y_train,c='b')
plt.show()
```

```python
test_pred = xgbr.predict(X_test)
plt.figure(figsize=(12,8))
X_axis = list(range(len(y_test)))
plt.plot(X_axis,test_pred,c='r')
plt.plot(X_axis,y_test,c='b')
plt.show()
```
